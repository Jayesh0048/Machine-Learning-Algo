```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("Iris.csv")
```

```python
df.info()
```

```python
df.head()
```

```python
X = df['SepalLengthCm'].values
y = df['PetalLengthCm'].values
```

```python
X
```

```python
y
```

```python
m = 0
b = 0
```

```python
lr = 0.01
n = len(X)
```

```python
epochs = 1000
```

```python
for step in range(epochs):

    # 1. Guess output using current line
    predicted_y = m * X + b

    # 2. Find how wrong the guesses are
    error = y - predicted_y

    # 3. Decide how to change slope
    slope_change = (-2/n) * np.sum(X * error)

    # 4. Decide how to move line up or down
    height_change = (-2/n) * np.sum(error)

    # 5. Update the line slowly
    m = m - lr * slope_change
    b = b - lr * height_change
```

```python
def predict(x):
    return m * x + b
```

```python
import matplotlib.pyplot as plt

plt.scatter(X, y)
plt.plot(X, m*X + b, color='red')
plt.show()
```
